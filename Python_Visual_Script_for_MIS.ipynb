```python
import os, uuid, matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot
import pandas

import sys
sys.tracebacklimit = 0

os.chdir(u'C:/Users/arpitag/PythonEditorWrapper_8f220e4d-d5d4-4e24-89f1-b0ccc27a4388')
dataset = pandas.read_csv('input_df_5ced0392-4e7f-4114-b50d-48cfda082e88.csv')

matplotlib.pyplot.figure(figsize=(5.55555555555556,4.16666666666667), dpi=72)
matplotlib.pyplot.show = lambda args=None,kw=None: matplotlib.pyplot.savefig(str(uuid.uuid1()))
# Original Script. Please update your script content here and once completed copy below section back to the original editing window #
import matplotlib.pyplot as plt
from matplotlib import patheffects
import pandas as pd
import numpy as np



# Assuming your dataset is already loaded and duplicates have been removed
dataset['Model'] = pd.Categorical(dataset['Model'])
dataset['ModelCode'] = dataset['Model'].cat.codes

unique_values = np.sort(dataset['YR-M'].unique())
value_to_number = {value: i for i, value in enumerate(unique_values)}
dataset['YR-M_scaled'] = dataset['YR-M'].map(value_to_number)

plt.figure(figsize=(30, 5))

# Create scatter plot for Call Count (renamed from Calls to YTD Call Count)
plt.scatter('YR-M_scaled', 'ModelCode', s='YTD Call Count', alpha=0.5, color='red', data=dataset, label='Call Count')
# Plotting production count as additional lighter grey bubbles
plt.scatter('YR-M_scaled', 'ModelCode', s='Production Count', alpha=0.1, color='green', data=dataset, label='Production Count')

# Define padding for y-axis limits
padding = 0.5
y_min, y_max = min(dataset['ModelCode'])-padding, max(dataset['ModelCode'])+padding
plt.ylim(y_min, y_max)

# Setting up yticks for the left y-axis
plt.yticks(ticks=range(len(dataset['Model'].cat.categories)), labels=dataset['Model'].cat.categories)
plt.gca().set_yticklabels(dataset['Model'].cat.categories, wrap=True)

# Create secondary y-axis to mirror the primary y-axis
secondary_ax = plt.gca().twinx()
# Set the same yticks for the secondary y-axis (right side)
secondary_ax.set_yticks(range(len(dataset['Model'].cat.categories)))
secondary_ax.set_yticklabels(dataset['Model'].cat.categories)
# Ensure the secondary y-axis has the same limits as the primary y-axis, including the padding
secondary_ax.set_ylim(y_min, y_max)

plt.ylabel('Model')
plt.xlabel('Appliance Year-Month')
plt.title('Model Performance Analysis')

# Adding data labels for call count number and percentage with only percentage bold and highlighted
for i, point in dataset.iterrows():
    call_count = f"{point['YTD Call Count']}"
    percentage = "0.0%" if point['Production Count'] == 0 else f"{point['YTD Call Count']/point['Production Count']*100:.1f}%"
    
    # Call count, not bold
    plt.text(point['YR-M_scaled'], point['ModelCode'], f"{call_count}|| ", 
             horizontalalignment='right', verticalalignment='center', fontsize=10)
    
    text_path_effects = [patheffects.withStroke(linewidth=3, foreground='white')]
    # Percentage, bold and highlighted
    percentage_text = plt.text(point['YR-M_scaled'], point['ModelCode'], f"{percentage}", 
             horizontalalignment='left', verticalalignment='center', fontsize=11, fontweight='bold')
    percentage_text.set_path_effects(text_path_effects)

# Production count labels and Delta value in bounding box
for i, point in dataset.iterrows():
    # Adjust this factor to scale the offset appropriately for Production number
    offset = np.sqrt(point['Production Count']) / 1000
    y_position = point['ModelCode'] + offset
    x_position = point['YR-M_scaled']
    
    # Production number
    plt.text(x_position, y_position, str(point['Production Count']), 
             horizontalalignment='center', verticalalignment='bottom', fontsize=10)

    # Adjust this factor to scale the offset appropriately for Delta
    delta_offset = np.sqrt(point['Production Count']) / 1000 
    y_position_delta = point['ModelCode'] - delta_offset 

    # Delta (Call Count) in a black box with white text
    delta_text = plt.text(x_position, y_position_delta, f" Î”: {point['Call Count']}", 
             bbox=dict(facecolor='black', alpha=0.7), color='white',
             horizontalalignment='center', verticalalignment='center', fontsize=10)

# Scale and format the x-axis
scaled_unique_values = [val for val in range(len(unique_values))]
plt.xticks(scaled_unique_values, unique_values, rotation=45, ha="right", fontsize=12)

plt.grid(True)
plt.show()
```
